```
import numpy as np
import random
```

#16題

```
temp=np.loadtxt("hw1_train.csv",dtype=np.float,delimiter='\t')
a=np.ones((1,100))
c = np.insert(temp, 0, values=a, axis=1)  #c是data b是sign
b=c[:,11]
c=np.delete(c, (11), axis=1)
updatelist=[]                  #updatelist是更新次數
for i in range(1000):
    correct=0                #correct為是否小於500
    update=0
    w=np.zeros((11))
    random.seed(i)
    while(correct<500):
        r=random.randint(0, 99)    #replacement
        if (np.dot(c[r].transpose(), w))*b[r]<=0:  
            w+=b[r]*c[r]
            correct=0
            update+=1
        else :
            correct+=1
    updatelist.append(update)
updatelist.sort()
print(updatelist[499],updatelist[500])
```
